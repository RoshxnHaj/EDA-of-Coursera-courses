### Loading data set

```python
import pandas as pd

def read_csv(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return None
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None

file_path = "C:/Users/KIIT/Desktop/DA project/CourseraDataset-Clean/CourseraDataset-Clean.csv"  # Update this with your file path
data = read_csv(file_path)

if data is not None:
    print("Dataset loaded successfully:")
    print(data.head())  # Display the first few rows of the dataset
```

### Plotting different courses offered by coursera based on the subject's frequency

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame and 'Keyword' is the column name
keyword_counts = data['Keyword'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
keyword_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Subjects')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
keyword_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.title('Distribution of Subjects')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
